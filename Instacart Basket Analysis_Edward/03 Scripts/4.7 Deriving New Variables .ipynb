```python
### Creating two new columns: two busiest days and two slowest days.
```

```python
ords_prods_merge['order_dow'].value_counts(dropna = False)
```

```python
result = []
sum_0_and_1 = 0
sum_3_and_4 = 4

for value in ords_prods_merge['order_dow']:
    if value == 0.0:
        result.append('Busiest days')
        sum_0_and_1 += 1
    elif value == 1.0:
        result.append('Busiest days')
        sum_0_and_1 += 1
    elif value == 4.0:
        result.append('Slowest days')
        sum_3_and_4 +3
    elif value == 3.0:
        result.append('Slowest days')
        sum_3_and_4 +3
    else: 
        result.append('Regularly busy')

```

```python
ords_prods_merge['busiest_days'] = result
```

```python
ords_prods_merge['busiest_days'].value_counts(dropna = False)
```

```python
ords_prods_merge['order_hour_of_day'].value_counts(dropna = False)
```

```python
#categorizing day into 3 parts Most orders, Average orders, and Fewest orders.
```

```python
result = []
sum_Most_orders = 10
sum_Average_orders = 17
sum_Fewest_orders = 23

for value in ords_prods_merge['order_hour_of_day']:
    if value in [9, 10, 11, 12, 13, 14, 15, 16]:
        result.append('Most orders')
        sum_Most_orders += 1  # Increment by 1 for each match
    elif value in [23, 6, 0, 1, 5, 2, 4, 3]:
        result.append('Fewest orders')
        sum_Fewest_orders += 1  # Increment by 1 for each match
    else: 
        result.append('Average orders')
        sum_Average_orders += 1  # Increment by 1 for all other matches
```

```python
ords_prods_merge['busiest_period_of_day'] = result
```
