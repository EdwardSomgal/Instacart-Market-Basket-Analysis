## 1. Renaming columns

```python
customers.rename(columns={'user_id': 'user_id',
                          'First Name': 'first_name',
                          'Surnam': 'last_name', 
                          'Gender': 'gender',
                          'fam_status': 'family_status',
                          'STATE': 'state',
                          'Age': 'age',
                          'number of dependants': 'dependants',
                          'income': 'income',
                          'date_joined': 'date_joined'
                                     
                                     }, inplace = True)
```

```python
customers.head()
```

## 2. Changing data type

```python
#changing data type of user_id to str
customers['user_id'] = customers['user_id'].astype('str')
```

## 3. Removing mix data type

```python
# Check for mix types
for col in customers.columns.tolist():
    weird = (customers [[col]].map(type) != customers[[col]].iloc[0].apply(type)).any(axis = 1)
    if len (customers[weird]) > 0:
        print(col)
```

```python
# changing datatype to string
customers['first_name'] = customers['first_name'].astype('str')
```

```python
# Cross Check for mix types
for col in customers.columns.tolist():
    weird = (customers [[col]].map(type) != customers[[col]].iloc[0].apply(type)).any(axis = 1)
    if len (customers[weird]) > 0:
        print(col)
```
