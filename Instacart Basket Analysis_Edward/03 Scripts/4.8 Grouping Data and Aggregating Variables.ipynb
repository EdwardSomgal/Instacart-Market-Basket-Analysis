### Department-wise average order 

```python
ords_prods_merge.groupby('department_id')['order_number'].mean()
```

```python
ords_prods_merge['Average_order'] = ords_prods_merge.groupby(['department_id'])['order_number'].transform(np.mean)
```

```python
# Average has a slight deviation as compared to the subset data, however, the biggest difference is with $2 for department_id 21. 
```

### Creating flags or categories of customers based on number of orders

```python
ords_prods_merge['Max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
ords_prods_merge.shape
```

```python
ords_prods_merge.loc[ords_prods_merge['Max_order'] > 40, 'loyalty_flag'] = 'Loyal Customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['Max_order'] <= 40) & (ords_prods_merge['Max_order'] > 10), 'loyalty_flag'] = 'Regular Customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['Max_order'] <= 10, 'loyalty_flag'] = 'New Customer'
```

```python
ords_prods_merge['loyalty_flag'].value_counts(dropna = False)
```

```python
ords_prods_merge.head()
```

```python
ords_prods_merge[['user_id', 'department_id', 'Max_order', 'loyalty_flag']].head(10)
```

```python
##Basic statistics product prices for each loyalty category 
```

```python
ords_prods_merge['mean_product_price'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
ords_prods_merge[['user_id', 'mean_product_price']].head(20)
```

### Types of Spenders

```python
# Create spending_flag column with given criteria using loc() function
ords_prods_merge.loc[ords_prods_merge['mean_product_price'] >= 10, 'spending_flag'] = 'High spender'
```

```python
ords_prods_merge.loc[ords_prods_merge['mean_product_price'] < 10, 'spending_flag'] = 'Low spender'
```

```python
#Check frequency of new spending_flag column
ords_prods_merge['spending_flag'].value_counts(dropna = False)
```

```python
# View relevant columns of updated data frame
ords_prods_merge[['user_id', 'mean_product_price', 'spending_flag']].head(20)
```

### Notification Plan

```python
ords_prods_merge['order_median'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
ords_prods_merge[['user_id', 'days_since_prior_order', 'order_median']].head(20)
```

```python
ords_prods_merge.loc[ords_prods_merge['order_median'] > 20, 'order_frequency'] = 'Non-frequent customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['order_median'] > 10) & (ords_prods_merge['order_median'] <= 20), 'order_frequency'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['order_median'] <= 10, 'order_frequency'] = 'Frequent customer'
```

```python
ords_prods_merge.head()
```

```python
# Check value counts of new order_frequency column
ords_prods_merge['order_frequency'].value_counts(dropna = False)
```

```python
# View relevant columns of updated data frame
ords_prods_merge[['user_id', 'days_since_prior_order', 'order_median', 'order_frequency']].head(25)
```
